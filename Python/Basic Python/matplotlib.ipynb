```python
#x-axis values
abc = [1,3,5,7,9]
#y-axis values
xyz=[10,9,5,7,8]
plt.plot(abc,xyz)
```

```python
plt.figure(figsize=(20,10)) #adjust the figure size
plt.bar(abc,xyz)
plt.barh(abc,xyz) #Horizantal
plt.show() #Basic editors need it to show the graph
```

```python
import numpy as np
year=[1993, 1995, 1997, 1999, 2001]
bangalore=[100.2, 111.5, 120.7, 130.9, 140.1]
mysore=[20.2, 30.5, 40.7, 50.9, 60.1]
plt.plot(year, bangalore, c='r', label="Bangalore", marker="*", linestyle='dashed'
, linewidth='3')
plt.plot(year, mysore, c='k', label="Mysor", marker=".")
plt.legend(loc="upper right")
plt.xlabel("Years", fontsize=10)
plt.ylabel("Emplyed People", fontsize=10)
plt.title("Emplyment Status", fontsize=10)
plt.xticks(np.arange(1993, 2012, 2), rotation=45)
plt.grid()
plt.show()
```
